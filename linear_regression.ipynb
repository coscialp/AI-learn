```python
from sklearn.datasets import make_circles, make_blobs, make_classification
import matplotlib.pyplot as plt

X, y = make_circles(n_samples=100, noise=0.1, factor=0.5)
# X, y = make_blobs(n_samples=1000, n_features=3, centers=2)
X = X.T
y = y.reshape((1, y.shape[0]))

print(X.shape)
print(y.shape)


plt.scatter(X[0, :], X[1, :], c=y)
plt.show()
```

```python
from neural_network import MLPClassifier
from sklearn.metrics import accuracy_score

model = MLPClassifier((100, 100, 100), n_iter=3000, learning_rate=0.04)

model.fit(X, y)

y_test = model.predict(X)

print(accuracy_score(y_test, y))
for epoch, loss in enumerate(model.loss_):
    print(f'epoch {epoch + 1}/{len(model.loss_)} - loss: {loss} - val_loss: {model.val_loss_[epoch]} - acc: {model.acc_[epoch]}')

model.display_train()
```
